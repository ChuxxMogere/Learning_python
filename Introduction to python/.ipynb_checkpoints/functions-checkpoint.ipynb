```python
# Familiar functions
# Find all of python functions in : https://docs.python.org/3/library/functions.html

var1 = [1, 2, 3, 4]
var2 = True

# print and type functions
print(type(var1))

# len functions to determine the length of something
print(len(var1))

# function to change the data type(type). var2 was initially a boolean but has been changed to integer
out2 = int(var2)
print(type(out2))
```

```python
# sorted function (arranges the list in asc or desc order)
first = [11.25, 18.0, 20.0]
second = [10.75, 9.50]

full = first + second

# Sort full in descending order. If you want in ascending order, reverse=False
full_sorted = sorted(full, reverse=True)
print(full_sorted)
```

```python
# METHODS (functions that belong to objects). There are many of them, this are but a few (do web search)
# string methods
place = "poolhouse"

# Use upper() method (To capitalize everything in the variable place)
place_up = place.upper()

# Print out place and place_up
print(place)
print(place_up)

# Using the .count() method to count number of o's
print(place.count("o"))

#Using .capitalize() method. Capitalizes the first letter
place2 = place.capitalize()
print(place2)

#Using .replace() method
place.replace("house", "room")
```

```python
# List methods
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Prints out the index of the element 20.0
print(areas.index(20.0))

# Print out how often 9.50 appears in areas
print(areas.count(9.50))
```

```python
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Use append twice to add poolhouse and garage size
areas.append(24.5)
areas.append(15.45)
print(areas)

# Reverse the orders of the elements in areas
areas.reverse()
print(areas)
```
