```python
# Python lists are a compound data type
hall = 11.25
kit = 18.0
liv = 20.0
bed = 10.75
bath = 9.50

areas = [11.25, 18.0, 20.0, 10.75, 9.50]
print(areas)
```

```python
#Adding strings to a list. See the name of the room and its area
areas = ["hallway", hall,
         "kitchen", kit,
         "living room", liv,
         "bedroom", bed,
         "bathroom", bath]
print(areas)
```

```python
# House information as list of lists (List inside a list
house = [["hallway", hall],
         ["kitchen", kit],
         ["living room", liv],
        ["bedroom", bed],
        ["bathroom", bath]]
print(house)
```

```python
#Subsetting lists (indexing)
house = ["hallway", 11.25, 
         "kitchen", 18.0, 
         "living room", 20.0, 
         "bedroom", 10.75, 
         "bathroom", 9.50]
# Print out second element from areas
print(house[1])

# Print out last element from areas
print(house[-1])

# Print out the area of the living room
print(house[5])
```

```python
# Use slicing to create downstairs
downstairs = house[0:6] #(The last index (6) is excluded

# Use slicing to create upstairs
upstairs = house[6:]

# Print out downstairs and upstairs
print(downstairs)
print(upstairs)
```

```python
#Lists inside a list
house = [["hallway", 11.25],
         ["kitchen", 18.0],
         ["living room", 20.0],
         ["bedroom", 10.75],
         ["bathroom", 9.50]]

# Subset the house list
print(house[-1])
print(house[-1][1])
```

```python
areas = ["hallway", 11.25, 
         "kitchen", 18.0, 
         "living room", 20.0, 
         "bedroom", 10.75, 
         "bathroom", 9.50]
# Update the area of the bathroom to be 10.50 square meters instead of 9.50 using negative indexing.
areas[-1]= 10.50
# Change "living room" to "chill zone"
areas[4]= "chill zone"
print(areas)
```

```python
# Extending a list
# Use the + operator to paste the list ["poolhouse", 24.5] to the end of the areas list. 
areas_1 = areas + ["poolhouse", 24.5]
print(areas_1)

# Further extend areas_1 by adding data on your garage
areas_2 = areas_1 + ["garage", 15.45]
print(areas_2)
```

```python
# Deleting list elements
areas = ["hallway", 11.25, "kitchen", 18.0,
        "chill zone", 20.0, "bedroom", 10.75,
         "bathroom", 10.50, "poolhouse", 24.5,
         "garage", 15.45]

# Delete the poolhouse items from the list
del areas[-4:-2]
print(areas)
```

```python
# Inner workings of lists
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change this command
areas_copy = areas

# Change areas_copy
areas_copy[0] = 5.0

# Print areas (Here, the both the areas and area copy change their index 0 to 5.0) 
print(areas)
```

```python
# Inner workings of lists
# However, sometimes when working on your list, 
#you'd want to have a copy and only make changes 
#to the copy and not the original list, thus you perform as below)
areas = [11.25, 18.0, 20.0, 10.75, 9.50]
areas_copy = list(areas)
areas_copy[0] = 5.0
print(areas)
print(areas_copy)
```
