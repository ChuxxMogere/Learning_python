```python
# Import a package
#calculating the area and circumference of a circle 
#using math package instead of writing the number of pi

import math

#Calculating the circumference and area
C = 2 * 0.43 * math.pi
A = math.pi * 0.43 **2

print("The circumference of the circle is: " + str(C))
print("The area of the circle is: " + str(A))
```

```python
# selective import(when you want to use a specific part of a package)
#import pi function from math package

from math import pi
C = 2 * 0.43 * pi
A = pi * 0.43 ** 2

print("Circumference is: "+ str(C))
print(f"Area is: {A:.3f}") #(writing in 3 dp)
```
